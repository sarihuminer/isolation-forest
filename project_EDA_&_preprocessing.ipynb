## Aggregations

Aggregation functions are operations that summarize or transform multiple values into a single value. They are used to compute summary statistics, such as mean, sum, count, minimum, maximum.

In Pandas, aggregation functions are commonly used in conjunction with the `groupby()` method to perform group-wise operations on DataFrames.

Basic aggregation fucntions:

1. `.sum()` - Sum of values
2. `.mean()` - Mean (average) of values
3. `.count()` - Count of non-null values
4. `.min()` - Minimum value
5. `.max()` - Maximum value
6. `.median()` - Median value
7. `.std()` - Standard deviation

## Window Functions and Rolling Windows

Window functions perform calculations across a set of rows related to the current row, known as a window. They are used to calculate statistics that involve a range of rows, such as moving averages, cumulative sums, or rankings.


Types of Window Functions in Pandas
1. Rolling Window Functions (`rolling`)
2. Expanding Window Functions (`expanding`)
3. Exponentially Weighted Window Functions (`ewm`)


### Rolling Window

```
df['column'].rolling(window, min_periods=None, center=False, win_type=None).aggregate_function()
```


1. `window`: Size of the moving window (number of observations).
2. `min_periods`: Minimum number of observations in window required to have a value.
3. `center`: Set the labels at the center of the window.



````python
df = pd.read_csv('flights_sample_3m.csv')
````

````python
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
df = df.sort_values(by='FL_DATE')
df.set_index('FL_DATE', inplace=True)
````

````python
df.head(15)
````

````python
df['7D_MA_ARR_DELAY'] = df['ARR_DELAY'].rolling(window=7).mean()


df[['ARR_DELAY', '7D_MA_ARR_DELAY']].head(15)
````

### Expanding Window Functions (expanding)


An expanding window includes all data from the beginning up to the current observation. It calculates cumulative statistics.



````python
df['CUMAVG_ARR_DELAY'] = df['ARR_DELAY'].expanding(min_periods=1).mean()

df[['ARR_DELAY', 'CUMAVG_ARR_DELAY']].head(15)
````

### Exponentially Weighted Window Functions (ewm)

Exponentially Weighted Moving functions apply weights that decay exponentially, giving more importance to recent observations.



```
df['column'].ewm(com=None, span=None, halflife=None, alpha=None, adjust=True).aggregate_function()
```

1. `com`: Center of mass
2. `span`: Specify decay in terms of span
3. `halflife`: Specify decay in terms of half-life
4. `alpha`: Smoothing factor directly

````python
df['EWMA_ARR_DELAY'] = df['ARR_DELAY'].ewm(span=7, adjust=False).mean()

df[['ARR_DELAY', 'EWMA_ARR_DELAY']].head(15)
````

### Choosing the Right Window Function
1. Rolling Window (`rolling`): Use when you need to smooth out short-term fluctuations and highlight longer-term trends or cycles.
2. Expanding Window (`expanding`): Use when you want to calculate cumulative statistics over time.
3. Exponentially Weighted Window (`ewm`): Use when you want to give more weight to recent observations in your calculations.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data with noise
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = x + np.random.normal(0, 1, 100)  # Add normal noise

# Calculate Pearson correlation coefficient
from scipy.stats import pearsonr
r, _ = pearsonr(x, y)
print(f'Pearson r = {r:.2f}')

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color='blue')
plt.title(f'Strong Positive Linear Relationship (Pearson r ≈ {r:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.savefig('scatter_plot_r=1.png')
plt.show()
````

````python
# Generate data with noise
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = -x + np.random.normal(0, 1, 100)  # Add normal noise

# Calculate Pearson correlation coefficient
r, _ = pearsonr(x, y)
print(f'Pearson r = {r:.2f}')

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color='red')
plt.title(f'Strong Negative Linear Relationship (Pearson r ≈ {r:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.savefig('scatter_plot_r=-1.png')
plt.show()
````

````python
# Generate data with no correlation
np.random.seed(0)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Calculate Pearson correlation coefficient
r, _ = pearsonr(x, y)
print(f'Pearson r = {r:.2f}')

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color='green')
plt.title(f'No Linear Correlation (Pearson r ≈ {r:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.savefig('scatter_plot_r=0.png')
plt.show()
````

````python
from scipy.stats import spearmanr

# Generate data
x = np.linspace(1, 1000, 1000)
y = np.power(x, 12) + np.random.normal(0, 10, 1000)  # Add noise

# Calculate correlation coefficients
pearson_r, _ = pearsonr(x, y)
spearman_r, _ = spearmanr(x, y)

print(f'Pearson r = {pearson_r:.2f}')
print(f'Spearman rho = {spearman_r:.2f}')

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, color='brown', alpha=0.6)
plt.title('Monotonic Decreasing Non-linear Relationship\n(Spearman Relevant, Pearson Less So)')
plt.xlabel('Variable X')
plt.ylabel('Variable Y = np.power(X, 12) + Noise')
plt.tight_layout()
plt.savefig('scatter_plot_spearman.png')
plt.show()
````
